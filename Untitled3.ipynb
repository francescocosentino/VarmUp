```python
from wordcloud import WordCloud, STOPWORDS


import matplotlib.pyplot as plt

bigstring = "GALE Digital Scholar Lab, text mining, reproducibility, data pipeline, clustering, sentiment analysis, wordcloud, corpus, data set, frequency analysis, filtering, big data"

plt.figure(figsize=(12,12))
wordcloud = WordCloud(stopwords=STOPWORDS,
                          background_color='white',
                          collocations=False,
                          width=1200,
                          height=1000
                         ).generate(bigstring)
plt.axis('off')
plt.imshow(wordcloud)
```
